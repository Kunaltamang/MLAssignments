```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

maximum = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
minimum = np.array( [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

time = np.arange(1,13,1)
tm = np.linspace(1,12,365)


def func(data,averg,ampl):
    return averg+ampl*np.cos(data*2*np.pi/data.max())

a,b  = curve_fit(func,time,maximum)
c,d = curve_fit(func,time,minimum)

plt.scatter(time,maximum)
plt.scatter(time,minimum)
plt.plot(tm,func(tm,*a))
plt.plot(tm,func(tm,*c))
plt.xlabel('Month')
plt.ylabel('Temperature(Â°C)')
plt.show()
```

```python
import pandas as pd
from collections import Counter

url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'

df = pd.read_csv(url)
sex = df['sex'][:-1]
cnt = Counter(sex)
key = cnt.keys()
val = cnt.values()

plt.pie(val,labels =key,autopct = '%.1f',startangle=90,radius=2)
plt.show()
```
